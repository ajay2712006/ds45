#Array properties

```python
import numpy as np
#1.shape
arr=np.arange(12)
print("Shape",arr.shape)

arr2=np.arange(12).reshape(3,4)
print("Shape",arr2.shape)

#2.dimension
print("Dimension",arr.ndim)
print("Dimension",arr2.ndim)

#3.size
print("Size",arr.size)
print("Size",arr2.size)

#4.dtype
print("Datatype",arr.dtype)
```
