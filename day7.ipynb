```python
#flatten()
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
flat_arr=arr.flatten()
print(flat_arr)


#for 3d
arr=np.arange(12).reshape(2,3,2)
print(arr)
flat_arr=arr.flatten()
print(flat_arr)
```

```python
#ravel-----> work on org value if we change a value in flat arr then it change org value
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
flat_arr=arr.ravel()
print(flat_arr)

#changing org value

flat_arr[0]=20
print("Original array")
print(arr)

#3d
arr=np.arange(12).reshape(2,3,2)
print(arr)
flat_arr=arr.ravel()
print(flat_arr)
```

#looping with numpy
1.while loop

2.for loop

```python
# while loop
# while loop for 1d

import numpy as np
arr = np.arange(12)
i = 0
while i < len(arr):
  print (arr[i], end=" ")
  i += 1
print()
print()
# while loop for 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
i=0
while i < len(arr):
  j=0
  while j < len(arr[i]):
    print(arr[i][j], end=" ")
    j += 1
  print()
  i += 1



```

```python
# while loop for 3d
arr = np.arange(24).reshape(2, 3, 4)

i = 0
while i < len(arr):
  j = 0
  while j < len(arr[i]):
    k = 0
    while k < len(arr[i][j]):
      print(arr[i][j][k], end=" ")
      k += 1
    print()
    j += 1
  print()
  i += 1
```

```python
#for loop
#print 1d array
import numpy as np
arr = np.arange(12)
for i in arr:
  print(i, end=" ")
print()
print()

#print 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
for i in arr:
  for j in i:
    print(j, end=" ")
  print()
print()


#print 3d array
import numpy as np
arr = np.arange(24).reshape(2,3,4)
for i in arr:
  for j in i:
    for k in j:
      print(k, end=" ")
    print()
  print()
```

#Statistical Functions
D
1. np.sum(a) -> Sum of all elements
く＞
2. np.mean(a) -> Average = Sum of elements / Number of elements
3. np.medilin(a) -> Middle value after sorting
4. np.min(a) -> Smallest value in array
5. np.max(a) -> Largest value in array
6. np.var(a) mean, difference, square, average | (sum = ddof) -> divide by array length

7. np.std(a) -> Standard Deviation = root(Variance) -> Measures spread of data
8. np.prod(a) -> Multiplication of all elements
9. np.cumsum(a) -> Cumulative (running) sum
10. np.cumprod(a) -> Cumulative (running) product
11. np.argmin(a) -> Index position of minimum value
12. np.argmax(a) -> Index position of maximum value
13. np.abs(a) -> Converts negative values to positive -> Absolute value (distance from zero)
14. np.unique(a) -> Returns unique values only -> Removes duplicates
15. np.percerftile(a, 50) -> Finds percentage-based value -> 50th percentile = Median
16. np.quantile(a, 0.5) -> Similar to percentile -> 0.5 = 50%

17. np.ptp(a) -> Range = Max - Min
田
18. np.any(a) -> True if at least one value is True

```python
# statistical Functions
#sum
import numpy as np
arr = np.array([2,3,5])
print(arr)
arr1 = np.sum(arr)

print("--------Add all elements-------- ")
print("sum : ",arr1)


#mean
import numpy as np
arr = np.array([2,3,5])
arr1 = np.mean(arr)
print("--------Average-------- ")
print("mean : ",arr1)

#median
import numpy as np
arr = np.array([2,3,4])
arr1 = np.median(arr)
print("--------Sort then find middle element-------- ")
print("median : ",arr1)

#min
import numpy as np
arr = np.array([2,3,5])
arr1 = np.min(arr)
print("--------Minimum element of array-------- ")
print("min : ",arr1)
#max
import numpy as np
arr = np.array([2,3,5])
arr1 = np.max(arr)
print("--------Maximum element of array-------- ")
print("max : ",arr1)
#var
import numpy as np
arr = np.array([2,3,5])
arr1 = np.var(arr)
print("--------Variance(avg of std)-------- ")
print("variance : ",arr1)
#std
import numpy as np
arr = np.array([2,3,5])
arr1 = np.std(arr)
print("--------Standard Deviation (2-mean) ka whole square similar for each element then there sum-------- ")
print("std : ",arr1)
#prod
import numpy as np
arr = np.array([2,3,5])
arr1 = np.prod(arr)
print("prod : ",arr1)
#cumsum
import numpy as np
arr = np.array([2,3,5])
arr1 = np.cumsum(arr)
print(arr1)
#cumprod
import numpy as np
arr = np.array([2,3,5])
arr1 = np.cumprod(arr)
print(arr1)
#argmin
import numpy as np
arr = np.array([2,3,5])
arr1 = np.argmin(arr)
print("argmin : ",arr1)
#argman
import numpy as np
arr = np.array([2,3,5])
arr1 = np.argmax(arr)
print("argmax : ",arr1)
#abs
import numpy as np
arr = np.array([2,3,5])
arr1 = np.abs(arr)
print(arr1)
#unique
import numpy as np
arr = np.array([2,3,5,2,3,5])
print(arr)
arr1 = np.unique(arr)
print(arr1)
#np.percentile(a,50)
import numpy as np
arr = np.array([2,3,5])
arr1 = np.percentile(arr,50)
print("percentile : ",arr1)
# np.quantile()
import numpy as np
arr = np.array([2,3,5])

arr1 = np.quantile(arr,0.5)
print("quantile : ",arr1)

#ptp
import numpy as np
arr = np.array([2,3,5])

arr1 = np.ptp(arr)
print("ptp : ",arr1)
#any
import numpy as np
arr = np.array([2,3,5])

arr1 = np.any(arr)
print("any : ",arr1)
```
