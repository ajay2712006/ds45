#Array Reshaping

1.reshape()

2.flatten() -> create copy

3.ravel() -> return original

4.transpose



```python
#reshape array
'''
import numpy as np
arr=np.arange(12)       #  -->12 can be in( 3,4)(6,2)(12,1)
up_arr=np.reshape(arr,(3,4))   #3-->row
print(up_arr)
'''

#example to make a 3d array
import numpy as np
arr=np.arange(12)
up_arr=np.reshape(arr,(2,3,2))
print(up_arr)
```

```python
#flatten()
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
flat_arr=arr.flatten()
print(flat_arr)


#for 3d
arr=np.arange(12).reshape(2,3,2)
print(arr)
flat_arr=arr.flatten()
print(flat_arr)
```

```python
#ravel-----> work on org value if we change a value in flat arr then it change org value
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
flat_arr=arr.ravel()
print(flat_arr)

#changing org value

flat_arr[0]=20
print("Original array")
print(arr)

#3d
arr=np.arange(12).reshape(2,3,2)
print(arr)
flat_arr=arr.ravel()
print(flat_arr)
```

```python
#transpose
import numpy as np
arr=np.arange(12).reshape(3,4)
print(arr)
trans_arr=arr.transpose()
print(trans_arr)

#for 3d
arr=np.arange(12).reshape(2,3,2)
print("3d array")
print(arr)
trans_arr=arr.transpose()
print(trans_arr)
```

```python
#slicing in 1d array
import numpy as np
arr=np.arange(11)
print(arr)
up_arr = arr[:3]
print(up_arr)

#for 2d
arr=np.arange(12).reshape(3,4)
print(arr)
up_arr=arr[:,-1]
print(up_arr)

#for print index 0 and 23 in 3d
arr=np.arange(24).reshape(2,3,4)     #2 table , 3row, 4col
print(arr)
up_arr = arr[0,0,0]
print(up_arr)
up_arr = arr[1,2,3]
print(up_arr)
```

#looping with numpy
1.while loop

2.for loop

```python
# while loop
# while loop for 1d

import numpy as np
arr = np.arange(12)
i = 0
while i < len(arr):
  print (arr[i], end=" ")
  i += 1
print()
print()
# while loop for 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
i=0
while i < len(arr):
  j=0
  while j < len(arr[i]):
    print(arr[i][j], end=" ")
    j += 1
  print()
  i += 1



```

```python
# while loop for 3d
arr = np.arange(24).reshape(2, 3, 4)

i = 0
while i < len(arr):
  j = 0
  while j < len(arr[i]):
    k = 0
    while k < len(arr[i][j]):
      print(arr[i][j][k], end=" ")
      k += 1
    print()
    j += 1
  print()
  i += 1
```

```python
#for loop
#print 1d array
import numpy as np
arr = np.arange(12)
for i in arr:
  print(i, end=" ")
print()
print()

#print 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
for i in arr:
  for j in i:
    print(j, end=" ")
  print()
print()


#print 3d array
import numpy as np
arr = np.arange(24).reshape(2,3,4)
for i in arr:
  for j in i:
    for k in j:
      print(k, end=" ")
    print()
  print()
```
