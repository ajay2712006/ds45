3. Guess a number

```python
import random

i = random.randint(1, 101)

user_num= int(input("Enter Number: "))
count = 1

while user_num != i:

    if i<user_num:
        print("Gusse lower")
        i=  random.randint(1,user_num)
    elif i>user_num:
        print("Gusse Greter")
        i =  random.randint(user_num,101)

    elif user_num == i:
        print(f"{i} is in the range")

    count +=1


print(f"You won in {count} Attempts")
```

4. print a table

```python
sec_no = 1
guess=int(input("Enter the number to print the table : "))
while (sec_no <= 10):
    print(f"{guess} x {sec_no} = {guess*sec_no}")
    sec_no += 1
```

5. check leap year or not

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```
