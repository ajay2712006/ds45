```python
import matplotlib.pyplot as plt

#
# **Markers**

# |character      |  |  |description |
# |-------------|  -------------------------------|
# |'.'       | | |point marker|
# |','       | | |pixel marker|
# |'o'       | | |circle marker|
# |'v'       | | |triangle_down marker|
# |'^'       | | |triangle_up marker|
# |'<'       | | |triangle_left marker|
# |'>'       | | |triangle_right marker|
# |'1'       | | |tri_down marker|
# |'2'       | | |tri_up marker|
# |'3'       | | |tri_left marker|
# |'4'       | | |tri_right marker|
# |'8'       | | |octagon marker|
# |'s'       | | |square marker|
# |'p'       | | |pentagon marker|
# |'P'       | | |plus (filled) marker|
# |'*'       | | |star marker|
# |'h'       | | |hexagon1 marker|
# |'H'       | | |hexagon2 marker|
# |'+'       | | |plus marker|
# |'x'       | | |x marker|
# |'X'       | | |x (filled) marker|
# |'D'       | | |diamond marker|
# |'d'       | | |thin_diamond marker|
# |'|'       | | |vline marker|
# |'_'       | | |hline marker|

# **Line Styles**

# |character      |  |  |  |description |
# |-------------|   -------------------------------|
# |'-'       | | | |solid line style|
# |'--'      | | | |dashed line style|
# |'-.'      | | | |dash-dot line style|
# |':'       | | | |dotted line style|

# Example format strings:

#     'b'    # blue markers with default shape
#     'or'   # red circles
#     '-g'   # green solid line
#     '--'   # dashed line with default color
#     '^k:'  # black triangle_up markers connected by a dotted line
# **Colors**

# |character      |  |  |  |color |
# |-------------|   -------------------------------|
# |'b'       | | | |blue|
# |'g'       | | | |green|
# |'r'       | | | |red|
# |'c'       | | | |cyan|
# |'m'       | | | |magenta|
# |'y'       | | | |yellow|
# |'k'       | | | |black|
# |'w'       | | | |white|


x = [2010,2015,2020,2025]    #x cord
y = [100,200,250,300]        #y cord
plt.plot(x,y,color="red",marker="o",linestyle="dashed",linewidth=4,markersize=14)
plt.xlabel("years")
plt.ylabel("sales")
plt.title("sales report")
plt.show()
```

```python
# multi bar chart
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y1 = [10,20,20,40]
y2 = [20,30,25,30]
# calculation -> width
w = 0.40
plt.bar(x - w/2,y1 , label="boys",width=w) # hide second
plt.bar(x + w/2,y2, label="girls",width=w) # show

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("Number of Students in Each group")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Categories
labels = ['Vehicle', 'Home', 'Work', 'Office']

# Data
x = np.arange(len(labels))

y1 = [1.0, 1.2, 2.3, 1.2]
y2 = [1.5, 1.4, 1.0, 3.2]
y3 = [4.0, 3.0, 2.0, 1.0]

# Width of bars
width = 0.2

# Plot bars
plt.figure(figsize=(8,5))

plt.bar(x - width, y1, width, label='Series 1', color='lightblue')
plt.bar(x,         y2, width, label='Series 2', color='steelblue')
plt.bar(x + width, y3, width, label='Series 3', color='lightgreen')

# Labels and ticks
plt.xticks(x, labels)
plt.ylabel('Values (Millions)')
plt.yticks([0,1,2,3,4], ['0','1M','2M','3M','4M'])

plt.title('Grouped Bar Chart')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])

y1 = [1.0, 1.2, 2.3, 1.2]
y2 = [1.5, 1.4, 1.0, 3.2]
y3 = [4.0, 3.0, 2.0, 1.0]

w = 0.20

plt.bar(x - w, y1, width=w, label="A")
plt.bar(x, y2, width=w, label="B")
plt.bar(x + w, y3, width=w, label="C")

plt.xticks(x, ["Vehicle", "Home", "Work", "Office"])

plt.legend()
plt.show()
```

```python
# First Chart
plt.subplot(1,2,1)

x = [1,2,3,4,5]
y = [10,20,30,40,55]

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Chart")

# Second Chart
plt.subplot(1,2,2)

x1 = ['apple','banana','orange','watermelon']
y1 = [40,30,15,70]

plt.pie(
    y1,
    labels=x1,
    autopct="%1.1f%%"
)

plt.title("Fruit Distribution")

plt.tight_layout()
plt.show()
```

#Sub plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,8))

# 1. Line Chart
plt.subplot(2,2,1)

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.title("Line Chart")


# 2. Bar Chart
plt.subplot(2,2,2)

x = ["A","B","C","D"]
y = [20,35,25,40]

plt.bar(x,y)
plt.title("Bar Chart")


# 3. Pie Chart
plt.subplot(2,2,3)

fruits = ["Apple","Banana","Orange","Mango"]
sales = [30,20,25,25]

plt.pie(sales, labels=fruits, autopct="%1.1f%%")
plt.title("Pie Chart")


# 4. Scatter Chart
plt.subplot(2,2,4)

x = [1,2,3,4,5]
y = [12,18,25,35,45]

plt.scatter(x,y)
plt.title("Scatter Chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,8))

# 1. Line Chart
plt.subplot(2,2,1)

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.title("Line Chart")


# 2. Bar Chart
plt.subplot(2,2,2)

x = ["A","B","C","D"]
y = [20,35,25,40]

plt.bar(x,y)
plt.title("Bar Chart")


# 3. Pie Chart
plt.subplot(2,2,3)

fruits = ["Apple","Banana","Orange","Mango"]
sales = [30,20,25,25]

plt.pie(sales, labels=fruits, autopct="%1.1f%%")
plt.title("Pie Chart")


# 4. Histogram
plt.subplot(2,2,4)

data = [10,12,15,18,20,22,25,28,30,35,40,45]

plt.hist(data, bins=5)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
