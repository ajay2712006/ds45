```python
#Lambda Functions [lambda arguments : expression]
add = lambda x: x
print (add (2000))

sum = lambda x,y: x+y
print (sum(10,20))
a = lambda x:x
print(a([10,20,30,40]))
```

```python
show = lambda *a: print(a)


show(1, 2, 3)
show("Ajay", "Hello")
```
