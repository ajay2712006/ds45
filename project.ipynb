```python
class Student:

    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks

    def display(self):
        print("Roll No:", self.roll_no)
        print("Name:   ", self.name)
        print("Marks:  ", self.marks)


students = []

while True:
    print("\n--- STUDENT SYSTEM ---")
    print("1. Add Student")
    print("2. Update Student")
    print("3. Delete Student")
    print("4. Search Student")
    print("5. View All Students")
    print("6. Exit")

    choice = input("\nEnter choice: ")

    if choice == "1":
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        marks = input("Enter Marks: ")

        s = Student(roll, name, marks)
        students.append(s)
        print("Student added successfully!")

    elif choice == "2":
        roll = input("Enter Roll No to update the info: ")
        found = False

        for s in students:
            if s.roll_no == roll:
                s.name = input("Enter new name: ")
                s.marks = input("Enter new marks: ")
                print("Record has been updated!")
                found = True
                break

        if found == False:
            print("Student is not in the list .")

    elif choice == "3":
        roll = input("Enter Roll No to delete : ")
        found = False

        for s in students:
            if s.roll_no == roll:
                students.remove(s)
                print("Student deleted .")
                found = True
                break

        if found == False:
            print("Student is not in the record .")

    elif choice == "4":
        roll = input("Enter Roll No of the student you want to find : ")
        found = False

        for s in students:
            if s.roll_no == roll:
                print("\n student Found :")
                s.display()
                found = True
                break

        if found == False:
            print("Student not found.")

    elif choice == "5":
        if len(students) == 0:
            print("No records available ,list is empty.")
        else:
            print("\n")
            print("All Students are : ")
            for s in students:
                s.display()

    elif choice == "6":
        break

    else:
        print("Invalid choice.")
```
