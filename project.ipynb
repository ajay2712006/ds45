```python
class Student:

    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks

    def display(self):
        print("Roll No:", self.roll_no)
        print("Name:   ", self.name)
        print("Marks:  ", self.marks)


students = []

while True:
    print("\n--- STUDENT SYSTEM ---")
    print("1. Add Student")
    print("2. Update Student")
    print("3. Delete Student")
    print("4. Search Student")
    print("5. View All Students")
    print("6. Exit")

    choice = input("\nEnter choice: ")

    if choice == "1":
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        marks = input("Enter Marks: ")

        s = Student(roll, name, marks)
        students.append(s)
        print("Student added successfully!")

    elif choice == "2":
        roll = input("Enter Roll No to update the info: ")
        found = False

        for s in students:
            if s.roll_no == roll:
                s.name = input("Enter new name: ")
                s.marks = input("Enter new marks: ")
                print("Record has been updated!")
                found = True
                break

        if found == False:
            print("Student is not in the list .")

    elif choice == "3":
        roll = input("Enter Roll No to delete : ")
        found = False

        for s in students:
            if s.roll_no == roll:
                students.remove(s)
                print("Student deleted .")
                found = True
                break

        if found == False:
            print("Student is not in the record .")

    elif choice == "4":
        roll = input("Enter Roll No of the student you want to find : ")
        found = False

        for s in students:
            if s.roll_no == roll:
                print("\n student Found :")
                s.display()
                found = True
                break

        if found == False:
            print("Student not found.")

    elif choice == "5":
        if len(students) == 0:
            print("No records available ,list is empty.")
        else:
            print("\n")
            print("All Students are : ")
            for s in students:
                s.display()

    elif choice == "6":
        break

    else:
        print("Invalid choice.")
```

```python
#Step 1: Create Dataset

import pandas as pd

data = {
    "Product":["Laptop","Mobile","Headphone","Laptop","Mobile",
               "Headphone","Laptop","Mobile","Headphone","Laptop"],
    "Category":["Electronics","Electronics","Accessories",
                "Electronics","Electronics","Accessories",
                "Electronics","Electronics","Accessories","Electronics"],
    "Month":["Jan","Jan","Jan","Feb","Feb",
             "Feb","Mar","Mar","Mar","Apr"],
    "Quantity":[5,10,15,4,8,12,6,11,10,7],
    "Price":[50000,20000,1500,50000,20000,
             1500,50000,20000,1500,50000]
}

df = pd.DataFrame(data)

#print(df)

#Step 2: Calculate Total Sales

df["Sales"] = df["Quantity"] * df["Price"]

#print(df)

#Step 3: NumPy Operations

import numpy as np

sales_array = np.array(df["Sales"])

print("Total Sales =", np.sum(sales_array))
print("Average Sales =", np.mean(sales_array))
print("Maximum Sale =", np.max(sales_array))
print("Minimum Sale =", np.min(sales_array))

#Step 4: Product-wise Sales

product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)

#Step 5: Category-wise Sales
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)

#Step 6: Monthly Sales

monthly_sales = df.groupby("Month")["Sales"].sum()

print(monthly_sales)

#Step 7: Bar Chart

import matplotlib.pyplot as plt

plt.bar(df["Product"], df["Sales"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product-wise Sales")
plt.show()

#Step 8: Pie Chart

category_sales.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Category Wise Sales")
plt.ylabel("")
plt.show()

#Step 9: Line Chart

monthly_sales.plot(
    kind="line",
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

#Step 10: Find Best Selling Product

best_product = product_sales.idxmax()

print("Best Selling Product =", best_product)
```
