#Save and Load data from CSV,json,excel O. url =
"https://raw.githubusercontent.com/rajendra0968jangid/machine_learning_demo/main/placement.csv"
1. csv import -> pd.read_csv ("filename")
2. csv export -> df.to_csv ('filename")
3. json import - > pd.read_json('filename')
4. json export -> df.to_json( filename')
5. excel import -> pd.read excel('filename', sheet name='Sheet1')
6. excel export -> df.to_excel('filename', index=False, sheet_name='Sheet1')

| Feature                  | CSV                    | Excel (XLSX)                   | JSON                               |
| ------------------------ | ---------------------- | ------------------------------ | ---------------------------------- |
| Full Form                | Comma-Separated Values | Microsoft Excel Workbook       | JavaScript Object Notation         |
| Data Structure           | Rows and columns       | Rows, columns, multiple sheets | Key-value pairs, nested objects    |
| Human Readable           | Yes                    | Yes (in Excel)                 | Yes                                |
| Supports Multiple Sheets | No                     | Yes                            | No                                 |
| Supports Formatting      | No                     | Yes (colors, fonts, charts)    | No                                 |
| Supports Formulas        | No                     | Yes                            | No                                 |
| File Size                | Small                  | Larger than CSV                | Medium                             |
| Used For                 | Simple tabular data    | Spreadsheets and reports       | Data exchange between applications |
| Extension                | `.csv`                 | `.xlsx`                        | `.json`                            |
